```python
WH_cols=['#','DateTime','Rain(mm)','Temp(C)','RH','WS','GS', 'DP']

#HOBO STATION 2019/2020 (uncomment for adding 2021 data and modify in a similar way for future years)

WH19=pd.read_table(r'WEATHER19.csv',sep=',',skiprows=2, names=WH_cols)
WH20=pd.read_table(r'WEATHER20.csv',sep=',',skiprows=2, names=WH_cols)
WH21=pd.read_table(r'WEATHER21.csv',sep=',',skiprows=2, names=WH_cols)

#this will merge both data sets continuously with the dates, it's important to do it before doing the indexing 
#WH=pd.concat([WH19,WH20], axis=0)
WH=pd.concat([WH19,WH20,WH21], axis=0)


#I drop the number column 
WH= WH.drop(columns=['#'])

#Create date time index and round times
WH.index = pd.to_datetime(WH['DateTime'])
WH.index=WH.index.floor('0.5H') # to round up time each 30 mins 

hobow=WH.resample('30min').ffill(limit=1) #forwardfill with a limit resamples and fills gaps with the next valid observation, we can limit to only fill one 
hobow=hobow.fillna(0)

hobow=hobow.drop(columns='DateTime')
#hobow.head()

#test for checking dframes 
#for exporting a csv file with the 'cleaned' data
#hobow.to_csv(r'/home/estefania/Documents/Backup_Enero_2021/Desktop/pruebaHOBO.csv', index = True)


```

```python
#I will need to add here and in the next box more radiation datasets 
RAD1=pd.read_table(r'RAD1.csv',sep=',',skiprows=1, names=R_cols)
RAD2=pd.read_table(r'RAD2.csv',sep=',',skiprows=1, names=R_cols)
RAD3=pd.read_table(r'RAD3.csv',sep=',',skiprows=1, names=R_cols)
RAD4=pd.read_table(r'RAD4.csv',sep=',',skiprows=1, names=R_cols)
RAD5=pd.read_table(r'RAD5.csv',sep=',',skiprows=1, names=R_cols)
RAD6=pd.read_table(r'RAD6.csv',sep=',',skiprows=1, names=R_cols)
RAD7=pd.read_table(r'RAD7.csv',sep=',',skiprows=1, names=R_cols)

#RAD1_in_pandas=RAD1.drop(columns=['Date,''Record'])#won't drop date so we can export

#rad1=RAD1_in_pandas.apply(lambda x: pd.to_numeric(x, errors='coerce')) 
#rad1=rad1.astype({'NetRad':'float','NetRadAvg':'float'}) 
#rad1.plot()
```

```python

RAD=pd.concat([RAD1,RAD2,RAD3,RAD4,RAD5,RAD6,RAD7], axis=0)
RAD=(RAD[~RAD['Date'].str.contains("[a-zA-Z]").fillna(False)]).dropna()
RAD=RAD.drop(columns=['Record'])
RAD.index = pd.to_datetime(RAD['Date'],errors='coerce')

IRR=RAD.apply(lambda x: pd.to_numeric(x, errors='coerce')) #changes argument to numeric type 
IRR=IRR.astype({'NetRad':'float','NetRadAvg':'float'}) #changes values to float 
IRR=IRR.clip(lower=0) #remove negative values 

rad=IRR.resample('30min').ffill(limit=1) #forwardfill with a limit resamples and fills gaps with the next valid observation, we can limit to only fill one 
rad=rad.fillna(0)

rad=rad.drop(columns='Date')

#len(rad)



#rad.to_csv(r'/home/estefania/Documents/Backup_Enero_2021/Desktop/pruebaRAD.csv', index = True)



#clever alternative method to get period but you need to know the final hour:
#di = date(y[0], m[0], d[0])
#df = date(y[-1], m[-1], d[-1])
#delta = df-di
#print(delta.days)
# p=delta.days*(24+(1/delta.days)) 
#the fraction added is to add an hour in the final set, I want a value such that x*total_no_ of_days is 1 h, If I want to add more hours I just do (tot no of h to add)/(tot no of days)
```

```python
hobow['date_only']=list(map(lambda index: index.date(), hobow.index)) 
tmin=hobow.groupby(['date_only']).min()
tmax=hobow.groupby(['date_only']).max()
#tmin
```

```python
hobow['TMIN']=[tmin[tmin.index == d]['Temp(C)'].values[0] for d in hobow.index]
hobow['TMAX']=[tmax[tmax.index == d]['Temp(C)'].values[0] for d in hobow.index]

```

```python
hobow['e0min']=0.6108*np.exp((17.27*hobow['TMIN'])/(237.3+hobow['TMIN']))

hobow['e0max']=0.6108*np.exp((17.27*hobow['TMAX'])/(237.3+hobow['TMAX']))
hobow.head()
```

```python
hobow['es']=(hobow['e0max']+hobow['e0min'])/2

#actual vapor pressure calculated by Temperature at Dew Point and by relative humidity they are pretty similar
hobow['ea1']=0.6108*np.exp((17.27*hobow['DP'])/(237.3+hobow['DP']))
hobow['ea2']=(hobow['RH']*(0.6108*np.exp((17.27*hobow['Temp(C)'])/(237.3+hobow['Temp(C)']))))/100
```

```python
hobow['VPD1']=hobow['es']-hobow['ea1']
hobow['VPD2']=hobow['es']-hobow['ea2']
```

```python
hobow['slope']=(2503*np.exp((17.27*hobow['Temp(C)'])/(237.3+hobow['Temp(C)'])))/((237.3+hobow['Temp(C)'])**2)
```

```python
hobowr=hobow.merge(rad['NetRad'], left_on=hobow.index, right_on=rad.index) #merge both hobo and radiation DF 
hobowr=hobowr.drop(columns='date_only')
#hobowr
```

```python
hobowr['PET1']=((0.408*hobowr['slope']*(hobowr['NetRad']*0.0864))+(0.000665*((900)/(hobowr['Temp(C)']+273))*(hobowr['WS']*hobowr['VPD1'])))/(hobowr['slope']+(0.000665*(1+0.34*hobowr['WS'])))

hobowr['PET2']=((0.408*hobowr['slope']*(hobowr['NetRad']*0.0864))+(0.000665*((900)/(hobowr['Temp(C)']+273))*(hobowr['WS']*hobowr['VPD2'])))/(hobowr['slope']+(0.000665*(1+0.34*hobowr['WS'])))
hobowr=hobowr.rename(columns={'key_0':'DateTime'})
hobowr.head()
```
