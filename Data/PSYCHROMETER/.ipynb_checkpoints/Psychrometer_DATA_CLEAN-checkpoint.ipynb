```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
#potply
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
```

```python
psymeters = [
    ('PSY1K30H.CSV', { #DF49
        'values' : 1,
        'values2': 1, 
        'output' : True
    }),
    ('PSY1J10I.CSV', { #ES48
        'values' : 1,
        'values2': 2,
        'output' : True 
    }),
    ('PSY1K30G.CSV', { #ES51
        'values' : 1,
        'values2': 5,
        'output' : True
    }),
    ('PSY1I604.CSV', { #DF21
        'values' : 1,
        'values2': 3,
        'output' : True
    }),
    ('PSY1I30Z-initial.CSV', { #ES50
        'values' : 2,
        'values2': 4,
        'output' : False
    }),
    ('PSY1J203-initial.CSV', { #DF27
        'values' : 2,
        'values2': 6,
        'output' : False
    })
    
]
```

```python
PSY_cols=['Date','Time','Chamber Temperature','dT','Wet Bulb Depression','Corrected Water Potential','Intercept','Slope','EDBO','Correction for dT','Correction Factor','IBV','IBT','EPS Present','EPS Voltage','EPS Current','Diagnostic Comment','NA'] #There is a file with something in the 18th column so we create an extra column
timeseries= []
```

```python
for i,(name,options) in enumerate(psymeters):
    if options['values'] == 1:
        #read file engine python is to avoid an error with the type of characters in the original file 
        PSYr=pd.read_csv(name,header = None, skiprows=1, names=PSY_cols, engine='python',encoding = "ISO-8859-1")
        #added the ,encoding = "ISO-8859-1" because otherwise it has an error on decoding UTF-8 

        PSY=(PSYr[~PSYr['Date'].str.contains("[a-zA-Z]").fillna(False)]).dropna(subset=['Date'])

        ind1 = pd.to_datetime(PSY['Date'] + ' ' + PSY['Time'],errors='coerce',format='%Y-%d-%m %H:%M:%S') 
        ind2 = pd.to_datetime(PSY['Date'] + ' ' + PSY['Time'],errors='coerce',format='%d/%m/%Y %H:%M:%S') 
 
        PSY.index=ind1.fillna(ind2)
        PSY=PSY.drop(PSY.loc[PSY.index.year<2019].index) 
        
    if options['values'] == 2:
        name2=name.rstrip('-initial.CSV')
        name_cont=name2 + ".CSV"
        PSY_i=pd.read_csv(name,header = None, skiprows=1, names=PSY_cols,engine='python',encoding = "ISO-8859-1")
        PSY_f=pd.read_csv(name_cont,header = None, skiprows=1, names=PSY_cols, engine='python',encoding = "ISO-8859-1")
        
        PSY_i=(PSY_i[~PSY_i['Date'].str.contains("[a-zA-Z]").fillna(False)]).dropna(subset=['Date'])
        PSY_f=(PSY_f[~PSY_f['Date'].str.contains("[a-zA-Z]").fillna(False)]).dropna(subset=['Date'])
        
       
        #same comments as above  except for this one, we are concattening 2 files because we have two separate files (replaced sd card) for these instruments 
        PSY=pd.concat([PSY_i,PSY_f], axis=0)
        
        ind1 = pd.to_datetime(PSY['Date'] + ' ' + PSY['Time'],errors='coerce',format='%Y-%d-%m %H:%M:%S') 
        ind2 = pd.to_datetime(PSY['Date'] + ' ' + PSY['Time'],errors='coerce',format='%d/%m/%Y %H:%M:%S') 
        PSY.index=ind1.fillna(ind2)
        PSY=PSY.drop(PSY.loc[PSY.index.year<2019].index)  
        
    if options['values2'] == 1:
        PSY=PSY.loc['2021-07-30 17:00:00':] #12/05/21 for 2021 with gaps/ In 2020 29/09/2020 14:30-06/11/2020:
        #print (PSY.index[0])
        a="DF49GT"
        
    if options['values2'] == 2:
        PSY=PSY.loc['2021-05-21 17:00:00':]
        a="ES48GT"
        
    if options['values2'] == 3:
        PSY=PSY.loc['2021-07-19 17:00:00':]
        a="DF21LS"
        
    if options['values2'] == 4:
        PSY=PSY.loc['2021-05-12 17:00:00':]
        a="ES50LS"
        
    if options['values2'] == 5:
        PSY=PSY.loc['2021-05-19 17:00:00':]
        a="ES51US"
        
    if options['values2'] == 6:
        PSY=PSY.loc['2021-05-18 17:30:00':]
        a="DF27US"
        
    
    #print(PSY)
        
    if options['output'] == True:
        dir_name= "/home/estefania/Documents/EstefaniaAcer_2021/Documents/PhD/Estefania data/FIELD DATA/Wasootch/2021/LATEST/PSYCHROMETER/CORRECTED/" # modify path and create a folder within working directory for CORRECTED files 
        import os
        nam=os.path.splitext(name)[0]
        outfile=nam + "-" + a + "-" + "corr" + ".csv"
        print (outfile)
        dir_name2= "/home/estefania/Documents/EstefaniaAcer_2021/Documents/PhD/Estefania data/FIELD DATA/Wasootch/2021/LATEST/PLOTS/CORRECTEDLOOP/" 
        
        psy_out=PSY.drop(columns=['Date','Time','Intercept','Slope','EDBO','Correction for dT','Correction Factor','IBV','IBT','EPS Present','EPS Voltage','EPS Current','Diagnostic Comment','NA'])
        psy_out.columns=['Chamber Temperature (C)','dT (C)','Wet Bulb Depression','Corrected Water Potential (MPa)']
        
        
        psy_out.to_csv(dir_name2 + outfile, index = True)
        psy_out.to_csv(dir_name + outfile, index = True)
        
       
        psy_ts=psy_out.reset_index()
        psy_ts['Water Potential (MPa)'] = psy_ts['Corrected Water Potential (MPa)'].astype(float)
        timeseries.append(psy_ts)

    if options['output'] == False:  
        import os
        dir_name= "/home/estefania/Documents/EstefaniaAcer_2021/Documents/PhD/Estefania data/FIELD DATA/Wasootch/2021/LATEST/PSYCHROMETER/CORRECTED/"
        nam=os.path.splitext(name2)[0] #use this function instead of rstrip to remove extnsion otherrwise I will lose S in SFM1I60S
        print (nam)
        #name2.rstrip(".CSV")
        outfile= nam + "-" + a + "-" + "corr" + ".csv"
        
        
        psy_out=PSY.drop(columns=['Date','Time','Intercept','Slope','EDBO','Correction for dT','Correction Factor','IBV','IBT','EPS Present','EPS Voltage','EPS Current','Diagnostic Comment','NA'])
        psy_out.columns=['Chamber Temperature (C)','dT (C)','Wet Bulb Depression','Corrected Water Potential (MPa)']
        
        psy_out.to_csv(dir_name + outfile, index = True)
        dir_name2= "/home/estefania/Documents/EstefaniaAcer_2021/Documents/PhD/Estefania data/FIELD DATA/Wasootch/2021/LATEST/PLOTS/CORRECTEDLOOP/"
        psy_out.to_csv(dir_name2 + outfile, index = True)
        
       
        
        psy_ts=psy_out.reset_index()
        psy_ts['Water Potential (MPa)'] = psy_ts['Corrected Water Potential (MPa)'].astype(float)
        timeseries.append(psy_ts)        
        
 
```
